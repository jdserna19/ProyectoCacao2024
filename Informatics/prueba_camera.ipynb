# Foto usando Matplotlib y OpenCV

Este programa permite capturar una imagen de una camara USB y graficarla usando la librer√≠a Matplotlib

## Librerias

```python
import cv2
import matplotlib.pyplot as plt
```

## Constantes

```python
NUM_CAMERA = 0 # Camara numero 0
```

## Capturar Imagen

```python
camera = cv2.VideoCapture(NUM_CAMERA)
return_value, image = camera.read()

# Liberar camara
del camera
```

## Mostrar imagen usando Matplotlib

```python
plt.imshow(image)
plt.title("Captured Image")
plt.axis("off")  # Hide axes
plt.show()
```

## Guardar imagen en disco

```python
cv2.imwrite('./pictures/imagen1.png', image)
```
